```python
texto = open("C:/Users/Diego/Desktop/TFM/datasets/todojunto_sub/set_train_and_validationV2/preparacion//train.en",encoding="utf8",errors='ignore')
```

```python
content=texto.readlines()
```

```python
dictionary = {}
```

```python
lista = []
```

```python
hyfon = "-"
hyfon2 = "--"
dot = "."
number1 = "1"
number2 = "2"
```

```python
array = [hyfon,hyfon2,dot,number1,number2]
```

```python
for line in content:
    line_aux = line.split()
    filtered_line = [word for word in line_aux if word not in stopwords.words('english')]
    filtered_line2 = [word for word in filtered_line if word not in array]
    if (hyfon in filtered_line2):
        if(filtered_line2.index(hyfon) == 0):
            filtered_line2.remove(hyfon)
    for word in filtered_line2:
        word_aux = word.lower()
        if (word_aux in lista):
            dictionary[word_aux] = dictionary[word_aux] + 1
        else:
            lista.append(word_aux)
            dictionary[word_aux] = 1
```

```python
for key,value in dictionary.items():
    tupl = [key,value]
    l.append(tupl)
```

```python
l2 = list(reversed(l))
```
